# Step 1: Import Libraries

```python
import numpy as np
from numpy import unique
from numpy import where
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from scipy.io import loadmat
from yellowbrick.cluster import KElbowVisualizer
```

# Step 2: Load and Process Data

```python
data = loadmat('Lab5_Data.mat')
print("Dimensions: ", len(data))
X = np.c_[data['X']]
```

The diminesions is 4.

```python
plt.scatter(X[:,0], X[:,1])
plt.xlabel('Exam-1 Score')
plt.ylabel('Exam-2 Score')
```

# Step 3: Train the Model

```python
# define the model
model = KMeans(n_clusters=2)
# fit the model
model.fit(X)
```

# Step 4: Use the Model

```python
# assign a cluster to each example
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)
```

```python
for cluster in clusters:
	# get row indexes for samples with this cluster
  row_ix = where(yhat == cluster)
	# create scatter of these samples
  plt.scatter(X[row_ix, 0], X[row_ix, 1])
  plt.xlabel('Exam-1 Score')
  plt.ylabel('Exam-2 Score')
  plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], marker='+', s=100, c='k', linewidth=2);
# show the plot
plt.show()
```

# Find Optimal K Using Elbow Method

```python
# define the model
model = KMeans(n_clusters=4)
# fit the model
model.fit(X)
# assign a cluster to each example
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)
for cluster in clusters:
	# get row indexes for samples with this cluster
  row_ix = where(yhat == cluster)
	# create scatter of these samples
  plt.scatter(X[row_ix, 0], X[row_ix, 1])
  plt.xlabel('Exam-1 Score')
  plt.ylabel('Exam-2 Score')
# show the plot
plt.show()
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,7))
# Fit the data to the visualizer
visualizer.fit(X)
# Finalize and render the figure
visualizer.show()
```
